```python
bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x <165 else('Average' if x <185 else 'Tall'))
#height category column using lambda function
```

```python
bios.head()
```

```python
#using functions to create a column and category of weight. 
def category_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'LightWeight'
    elif row['height_cm'] < 185 and row['weight_kg'] <= 80:
        return 'MiddleWeight'
    else:
        return 'HeavyWeight'
    
bios['weight_category'] = bios.apply(category_athlete, axis=1)  #axis 1 = row and axis=0 is columns. 
bios.head()
```

Merging andconcanating the dataframes. 

```python
bios2 = bios_new.copy()
```

```python
bios2 = pd.merge(bios2, nocs, left_on='born_country', right_on='NOC', how='left')
bios2.head(10)

#Joining the tables with left join. 
# After HOW='LEFT' we can add suffix to give name to seprate the columns: suffixes=''. 
```

```python
bios2.drop(['NOC_y', 'region_x','notes_x'], axis=1, inplace=True)
#dropping a column (multiple in this case)
```

```python
bios2.head(10)
```

```python
bios2.rename(columns={'NOC_x':'NOC_bios'}, inplace = True)
bios2.rename(columns={'region':'born_country_full'}, inplace=True)
bios2.head(10)
#rename the column in tha df. 
```

```python
#find the names wehre the olympic country is iff and bon country is different. . 
bios2[bios2['NOC_bios'] != bios2['born_country_full']][['name','NOC_bios','born_country_full']]
```

```python
usa = bios2[bios2['born_country_full'] == 'USA'].copy()
gbr = bios2[bios2['born_country_full'] == 'GBR'].copy()
```

```python
#this is simple concate function use case in pandas. the first table USA is on top of GBR
new_df = pd.concat([usa, gbr])
new_df.head(10)
```
